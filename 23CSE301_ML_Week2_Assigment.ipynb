```python
#Load the dataset into a dataframe

import pandas as pd
df=pd.read_csv("./Vehicle_Registrations.csv")
```

```python
#Display the first 10 and last 5 rows
df.head(10)
```

```python
#Display the last 5 rows
df.tail(5)
```

```python
#Show summary statistics and data types of each column.

for name, dtype in df.dtypes.items():
    print(f"Column: {name}, Data Type: {dtype}")
      
```

```python
#Filter records based on a condition (e.g., values greater than a threshold).

df[df['State Name (state_name)']=='Tamil Nadu']
```

```python
#Add a new derived column using existing columns.

def state_classification(state_name):
    south_India = ["Tamil Nadu", "Kerala", "Karnataka", "Andhra Pradesh", "Telangana", "Goa"]
    west_India = ["Maharashtra", "Gujarat"]
    east_India = ["Odisha", "Chhattisgarh", "Jharkhand", "Bihar", "West Bengal", "Sikkim"]
    northeast_India = ["Assam", "Manipur", "Meghalaya", "Tripura", "Mizoram", "Nagaland", "Arunachal Pradesh"]
    northcentral_India = ["Rajasthan", "Madhya Pradesh", "Uttar Pradesh", "Uttarkhand", "Haryana"]
    north_India = ["Punjab", "Himachal Pradesh"]
    union_Territory = [
        "Delhi", "Puducherry", "Lakshadweep", "Andaman and Nicobar Islands",
        "Jammu and Kashmir", "Ladakh", "Dadra and Nagar Haveli and Daman and Diu", "Chandigarh"
    ]

    india = {
        "South India": south_India,
        "West India": west_India,
        "East India": east_India,
        "Northeast India": northeast_India,
        "Northcentral India": northcentral_India,
        "North India": north_India,
        "Union Territory": union_Territory
    }

    for region in india:
        if state_name in india[region]:
            return region

df["Region Type "]  =df["State Name (state_name)"].apply(state_classification)      
 
df.head()      
```

```python
#Group the data by a categorical column and compute mean/median for another numeric column.

#For this, we will group the data by state names

df.groupby('State Name (state_name)').mean(numeric_only=True)
```

```python
#Sort the DataFrame based on one or more columns.
grouped_df=df.groupby('State Name (state_name)').mean(numeric_only=True)
new_df=grouped_df.sort_values('Registrations (registrations)', ascending=True)
```

```python
#Handle missing values by either dropping or filling them.
new_df.isnull().sum()
```

```python
#Export the final DataFrame to a new CSV file.

df["Region Type "]=df["State Name (state_name)"].apply(state_classification)  
df.to_csv('Analysed_DataSet.csv', index=False)
```

```python
#One distribution plot (histogram, KDE, or violin plot).
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution plot of a numeric column
sns.histplot(df['Registrations (registrations)'], bins=30, kde=True)

plt.title("Distribution of Vehicle Registrations")
plt.xlabel("Registrations")
plt.ylabel("Frequency")
plt.show()
```

```python
#One categorical plot (barplot, countplot, or boxplot).

df.rename(columns=lambda x: x.strip(), inplace=True)

sns.barplot(x='Region Type', y='Registrations (registrations)', data=df, estimator='mean')
plt.title("Average Vehicle Registrations per Region")
plt.xticks(rotation=45)
plt.show()
```

```python
#One relationship plot (scatterplot or lineplot).

import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'Date (date)' is in datetime format
df['Date (date)'] = pd.to_datetime(df['Date (date)'])

sns.scatterplot(x='Date (date)', y='Registrations (registrations)', data=df)

plt.title("Vehicle Registrations Over Time")
plt.xlabel("Date")
plt.ylabel("Registrations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#A correlation heatmap of numerical columns.

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Compute correlation matrix (only numeric columns)
corr_matrix = df.corr(numeric_only=True)

# 2. Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```

```python
#A pairplot to compare multiple relationships.

import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot of all numeric columns
sns.pairplot(df)

plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```
